# Dictionary
A dictionary is a key:value collection. </p>
Each key has a value, so each value is "named" </p>
**Syntax:** {} -> empty dictionary </p>
{ key: value} -> dictionary with one key:value </p>
{ key:value, key2:value2, ...} </p>
**Considerations:** </p>
The key is always a string, the value can be of any type </p>
The key is always unique </p>

```python
empty_dictionary = {} # empty dictionary
dictionary = {"name": "Anna"}

print(dictionary)
harrys_list = ['no clue', 'one clue', 'two clues', 'many clues']

# Amsterdam NLD Noord-Holland 731200
amsterdam = {
    "city_name": "Amsterdam",
    "country_code": "NLD",
    "district": "Noord-Holland",
    "population": 731200,
    "potential_clues": harrys_list
  }

print(amsterdam)

print(amsterdam['population']) # select one value
print(amsterdam['potential_clues'][-1])

# replace a value in a dictionary
amsterdam['population'] = 731201
print(amsterdam['population'])

# add a key:value to a dictionary
amsterdam['language'] = 'Dutch'
print(amsterdam)
print(amsterdam['language'])
```

## While loop
